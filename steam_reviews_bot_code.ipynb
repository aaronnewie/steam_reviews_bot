# Libraries and Imports


```python
# Libraries for dataframes
%pip install steamreviews
import steamreviews
import pandas as pd
import re

# Libraries for correlation
import numpy as np
from scipy import stats

# Libraries for plotting
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import time

# Libraries for text analysis
%pip install -U textblob
from textblob import TextBlob
```

# Chatbot

```python
# Checks to see if user inputed a proper link.
# Uses re.findall to extract app_id from steam_url only.
link_check = True
while link_check == True:
    try:
        steam_url = input('''Hi, my name is Steambot! I can analyze any store
link from Steam! Please copy and paste your link below to get started!''')
        regex = re.findall('\d+', steam_url)
        app_id = regex[0]
        link_check = False
    except:
        steam_url = input('''Oops! Please insert a storelink by going
to the steam webstore!: ''')
        regex = re.findall('\d+', steam_url)
        app_id = regex[0]
        link_check = False

# Gathers steamreviews API data      
request_params = dict()
# Gathers steamreviews in english only
request_params['language'] = 'english'
print('Gathering API data.... This might take a bit!')
review_dict, query_count = steamreviews.download_reviews_for_app_id(app_id,
                           chosen_request_params=request_params)

# Creates base Dataframe from app_id based on review information
app_id = regex[0]
pd.set_option("display.max_rows", None, "display.max_columns",
              None, "display.max_colwidth", None)
steam_df = pd.DataFrame.from_dict(review_dict['reviews'], orient = 'index',
                                  columns = ['votes_up','votes_funny',
                                             'voted_up','review'])
steam_df.reset_index(inplace = True)
steam_df.dropna(inplace = True)

# Dataframe for positive reviews only 
steam_positive_df = steam_df[steam_df.voted_up == True]
# Dataframe for most upvoted positive reviews
max_positive = steam_positive_df.votes_up.argmax()
# Dataframe for negative reviews only
steam_negative_df = steam_df[steam_df.voted_up != True]
# Dataframe for most upvoted negative reviews
max_negative = steam_negative_df.votes_up.argmax()

# Dataframe for correlation analysis
correlation_df = stats.pearsonr(steam_df['votes_up'], steam_df['votes_funny'])

# Most upvoted positive review sentiment analysis function
def positive_sentiment_analysis(index):
    """Creates a dataframe for based on the index of the highest upvoted
     positive review. And, uses TextBlob for sentiment analysis"""
    steam_df = pd.DataFrame.from_dict(review_dict['reviews'],
                                              orient = 'index', columns =
                                              ['votes_up', 'voted_up',
                                               'review'])
    steam_positive_df = steam_df[steam_df.voted_up == True]
    max_positive = steam_positive_df.votes_up.argmax()
    review = steam_positive_df['review'][max_positive]
    text = review
    analysis = TextBlob(text).sentiment
    print(text)
    print(analysis)
    return
    
# Most upvoted negative reiew sentiment analysis function 
def negative_sentiment_analysis(index):
    """Creates a dataframe for based on the index of the highest upvoted
     negative review. And, uses TextBlob for sentiment analysis"""
    steam_df = pd.DataFrame.from_dict(review_dict['reviews'],
                                              orient = 'index', columns =
                                              ['votes_up', 'voted_up',
                                               'review'])
    steam_negative_df = steam_df[steam_df.voted_up != True]
    max_negative = steam_negative_df.votes_up.argmax()
    review = steam_negative_df['review'][max_negative]
    text = review
    analysis = TextBlob(text).sentiment
    print(text)
    print(analysis)
    return

# Correlation function between votes_up and votes_funny    
def cor_calc(correlation_df):
    """Uses ScyPy to create a data frame between positive votes and funny votes
     to perform correlational analysis"""
    steam_df = pd.DataFrame.from_dict(review_dict['reviews'], 
    orient = 'index', columns = ['votes_up', 'votes_funny', 'review'])
    linear_relationship = correlation_df[0]
    p_value = correlation_df[1] 
    print(correlation_df)

# four different outcomes for correlational analysis   
    if linear_relationship > 0 and p_value <= 0.05:
        print('''A positive correlation can exist and 
since the p-value is less than 
or equal to 0.05 we can reject the null hypothesis.''')
    elif linear_relationship < 0 and p_value <= 0.05:
        print('''A negative correlation can exist and 
since the p-value is less than 
or equal to 0.05 we can reject the null hypothesis.''')
    elif linear_relationship > 0 and p_value >= 0.05:
        print('''A positive correlation can exist however 
since the p-value is greater than 
or equal to 0.05 we can not reject the null hypothesis.''')
    elif linear_relationship < 0 and p_value >= 0.05:
        print('''A negative correlation can exist however 
since the p-value is greater than 
or equal to 0.05 we can not reject the null hypothesis.''')

# Creates Dataframe based on overall reviews for bargraph
review_dict['query_summary']
ratings_df = pd.DataFrame.from_dict(review_dict['query_summary'],
                                    orient = 'index', columns = ['Values'])
ratings_df.reset_index(inplace = True)
ratings_df.drop(index = [0, 1, 2 , 5], inplace = True)

# Chatbot branch starts from here
again = True
while again == True:
    print('''Please input the words in brackets!: ''')
    ror = input('''Would you like to see data on the overall [ratings]
or [reviews]?: ''')
    ror_lower = ror.lower()
    while ror_lower not in ['ratings', 'reviews']:
        print('Please only type [ratings] or [reviews] in the box!')
        ror = input('''Would you like to see data on
the overall [ratings] or [reviews]?: ''')
        ror_lower = ror.lower()

# Overall Ratings Bargraph - taken from ratings_df
    if ror_lower == 'ratings':
        print('Let\'s check out a bargraph then!')
        ratings_bar = sns.barplot( x = ratings_df['index'],
                               y = ratings_df['Values'])
        plt.show() 
        time.sleep(.5) 
        response = input('''Would you like to see more information?
Input [no] to stop: ''')
        if response == 'no':
            again = False
            print('Thanks for using Steambot!')
    
# Start of Reviews Branch
    elif ror_lower == 'reviews':
        review_selection = input('''Please choose from [positive], [negative], 
[funny], or [random]: ''')
        review_selection_lower = review_selection.lower()
        while review_selection_lower not in ['positive', 'negative', 'funny',
                                             'random']:
                review_selection = input('''Please choose from [positive],
[negative], [funny], or [random]: ''')
                review_selection_lower = review_selection.lower()
                
# Positive Reviews
        if review_selection_lower == 'positive':
            positive_review = input('''Would you like to see the [top 5]
reviews or [sentiment analysis] of the most upvoted positive review?: ''')
            positive_review_lower = positive_review.lower()
            while positive_review_lower not in ['top 5', 'sentiment analysis']:
                print('Please enter [top 5] or [sentiment analysis] ')
                positive_review = input('''Would you like to see the [top 5]  
reviews or [sentiment analysis] of the most upvoted positive review?: ''')
                positive_review_lower = positive_review.lower()
            if positive_review_lower == 'top 5':
                print('Displaying top 5 positively rated reviews. ')
                display(steam_positive_df.nlargest(5, 'votes_up'))
                response = input('''Would you like to see more information?
Input [no] to stop: ''')
                if response == 'no':
                    again = False
                    print('Thanks for using Steambot!')
            elif positive_review_lower == 'sentiment analysis':
                print('Displaying sentiment analysis.')
                positive_sentiment_analysis(max_positive)
                response = input('''Would you like to see more information?
Input [no] to stop:''')
                if response == 'no':
                    again = False
                    print('Thanks for using Steambot!')

# Negative Reviews
        elif review_selection_lower == 'negative':
            negative_review = input('''Would you like to see the [top 5]  
reviews or [sentiment analysis] of the most upvoted negative review?: ''')
            negative_review_lower = negative_review.lower()
            while negative_review_lower not in ['top 5', 'sentiment analysis']:
                print('Please enter top 5 or sentiment analysis')
                negative_review = input('''Would you like to see the [top 5]
reviews or [sentiment analysis] of the most upvoted negative review?: ''')
                negative_review_lower = negative_review.lower()
            if negative_review_lower == 'top 5':
                print('Displaying top 5 negatively rated reviews. ')
                display(steam_negative_df.nlargest(5, 'votes_up'))
                response = input('''Would you like to see more information?
Input [no] to stop: ''')
                if response == 'no':
                    again = False
                    print('Thanks for using Steambot!')
            elif negative_review_lower == 'sentiment analysis':
                print('Displaying sentiment analysis.')
                negative_sentiment_analysis(max_negative)
                response = input('''Would you like to see more information?
Input [no] to stop: ''')
                if response == 'no':
                    again = False
                    print('Thanks for using Steambot!')

# Funny Reviews (Based on votes_funny)
        elif review_selection_lower == 'funny':
            funny_review = input('''[Top 5] or [Correlation]?: ''')
            funny_review_lower = funny_review.lower()
            while funny_review_lower not in ['top 5', 'correlation']:
                print('Please enter correlation or top 5: ')
                funny_review = input('''[Top 5] or [Correlation]?: ''')
                funny_review_lower = funny_review.lower()
            if funny_review_lower == 'top 5':
                print('Displaying top 5 funniest reviews. ')
                display(steam_df.nlargest(5, 'votes_funny'))
                response = input('Try again?: Input [no] to stop: ')
                if response == 'no':
                    again = False
                    print('Thanks for using Steambot!')
            elif funny_review_lower == 'correlation':
                print('''Displaying the correlation between positive and
funny reviews.''')
                cor_calc(correlation_df)
                response = input('''Would you like to see more information?
Input [no] to stop: ''')
                if response == 'no':
                    again = False
                    print('Thanks for using Steambot!')

        elif review_selection_lower == 'random':
            print('Displaying 5 random reviews from the game of your choice!')
            display(steam_df.sample(n = 5))
            response = input('''Would you like to see more information?
Input [no] to stop: ''')
            if response == 'no':
                again = False
                print('Thanks for using Steambot!')
```
